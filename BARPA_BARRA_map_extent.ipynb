```python
import xarray as xr
from plotting_maps.acs_plotting_maps import plot_acs_hazard_multi, plot_acs_hazard, plot_data, cmap_dict, regions_dict
from matplotlib import colors, cm
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
```

```python
ds = xr.open_dataset("/scratch/v46/gt3409/windspeed/200hPa/freq-above-p99/windspeed-200hPa-monthly-freq-above-p99_AUS-15_historical_ACCESS-CM2_r4i1p1f1_BOM_BARPA-R_v1-r1_6hr.nc")
ds
```

```python
ds_mean = ds["windspeed"].mean("time").compute()
```

```python
plt.figure(figsize=(12,6))
ax = plt.subplot(projection=ccrs.PlateCarree(180))
ds_mean.plot( transform=ccrs.PlateCarree(), ax=ax)
plt.title("")
ax.yaxis.set_visible(True)
ax.xaxis.set_visible(True)
# ax.grid()
ax.set_xlim(-180, 180)
ax.set_ylim(-90, 90)
ax.set_xticks(np.arange(-180, 181, 30))
ax.set_xticklabels(np.arange(0,361,30))
ax.coastlines()

#BARPA
x_min = 86.336
x_max = 209.4725
y_min = -55.275
y_max = 15.331

rect = mpatches.Rectangle(
    (x_min, y_min), x_max-x_min, y_max-y_min,
    edgecolor='red',
    facecolor="none",
    alpha=1,
    transform=ccrs.PlateCarree(0))
ax.add_patch(rect)

#BARRA
y_min, y_max = (-65.0, 19.48)
x_min, x_max = (65, 196.89)
rect = mpatches.Rectangle(
    (x_min, y_min), x_max-x_min, y_max-y_min,
    edgecolor='red',
    facecolor="none",
    ls="--",
    alpha=1,
    transform=ccrs.PlateCarree(0))
ax.add_patch(rect)
```
